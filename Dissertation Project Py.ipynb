## Adding new columns  males, female, below 65, between 65-75, over 75

```python
cox_fd = pd.read_csv("cox_fd.csv")
```

```python
cox_fd
```

```python

for index, row in cox_fd.iterrows():
    if(cox_fd.loc[index,'ragender'] == 1):
       cox_fd.loc[index,'male'] = 1
    else:
       cox_fd.loc[index,'male'] = 0
    if(cox_fd.loc[index,'ragender'] == 2):
       cox_fd.loc[index,'female'] = 1
    else:
       cox_fd.loc[index,'female'] = 0
    if(cox_fd.loc[index,'age_vc'] == 1):
       cox_fd.loc[index,'below_65'] = 1
    else:
       cox_fd.loc[index,'below_65'] = 0
    if(cox_fd.loc[index,'age_vc'] == 2):
       cox_fd.loc[index,'between_65_75'] = 1
    else:
       cox_fd.loc[index,'between_65_75'] = 0
    if(cox_fd.loc[index,'age_vc'] == 3):
       cox_fd.loc[index,'over_75'] = 1
    else:
       cox_fd.loc[index,'over_75'] = 0
```
